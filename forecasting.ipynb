```python
import pandas as pd

# Replace with the correct CSV file name inside the zip
data = pd.read_csv("dataset/RETAILSMNSA.csv")

print(data.head())
```

```python
print(data.head())
print(data.info())
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
print(data.head())
print(data.info())
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data = data.groupby('date').sum().reset_index()
print(data)
```

```python
print(data.columns)
```

```python
data = data.rename(columns={'value': 'sales'})
data = data[['date', 'sales']]   # keep only date and sales
data['date'] = pd.to_datetime(data['date'])
```

```python
print(data.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(data['date'], data['sales'])
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
from prophet import Prophet

# Prophet needs columns named 'ds' and 'y'
df = data.rename(columns={'date':'ds', 'sales':'y'})

model = Prophet()
model.fit(df)

# Forecast for next 30 days
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

# Plot forecast
model.plot(forecast)
plt.show()

# Plot seasonality/trend
model.plot_components(forecast)
plt.show()
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
forecast.to_csv("sales_forecast.csv", index=False)
```
